```python
import requests, json, time
import pandas as pd
import numpy as np
import warnings; warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv("output5.csv", names = ['제목', '장르', '평점', '투표수', '사이트'])
print(movies.shape)
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
movies['장르'] = movies['장르'].apply(lambda x : x.replace("[", ''))
movies['장르'] = movies['장르'].apply(lambda x : x.replace("]", ''))
movies['장르'] = movies['장르'].apply(lambda x : x.replace(",", ''))
movies['장르'] = movies['장르'].apply(lambda x : x.replace("'", ''))
```

```python
movies
```

```python
count_vect = CountVectorizer(min_df = 0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(movies['장르'])
print(genre_mat.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim)
```

```python
genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_ind[:1])
```

```python
C = movies['평점'].mean()
m = movies['투표수'].quantile(0.6)

def weighted_vote_average(record):
    v = record['투표수']
    R = record['평점']
    
    return ((v/(v+m))*R) + ((m/(m+v))*C)

movies['투표가중치']=movies.apply(weighted_vote_average, axis=1)
```

```python
movies[['제목', '평점', '투표수', '투표가중치']].sort_values('투표가중치', ascending = False)[:10]
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    
    title_movie=df[df['제목']==title_name]
    title_index=title_movie.index.values
    similar_indexes=sorted_ind[title_index, :(top_n)]
    
    similar_indexes=similar_indexes.reshape(-1)
    similar_index = similar_indexes[similar_indexes!=title_index]
   
    return df.iloc[similar_indexes].sort_values('투표가중치', ascending=False)[:top_n]
```
